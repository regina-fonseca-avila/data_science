````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

# K-means
## Como funciona

````python
# conjunto de  pontos para exemplo
x = [1,2,3.4,4.5,9,10,11,12,15,15.5,16,17]
y = list(np.zeros(12))
````

## Quero agrupar um conjunto de dados em 3 clusters

````python
plt.figure(figsize=(20,10))
plt.scatter(x, y, c = 'grey', s = 250, zorder = 2)
plt.ylim(-0.2,0.3)
plt.axhline(c = 'black', zorder = 1)
plt.axis('off');
````

## Intuitivamente podemos agrup√°-los assim

````python
plt.figure(figsize=(20,10))
plt.scatter(x[0:4], y[0:4], c = 'blue', s = 250, zorder = 2)
plt.scatter(x[4:8], y[4:8], c = 'green', s = 250, zorder = 2)
plt.scatter(x[8:12], y[8:12], c = 'orange', s = 250, zorder = 2)
plt.ylim(-0.2,0.3)
plt.axhline(c = 'black', zorder = 1)
plt.axis('off');
````

## E usando o K-means?

### 1. Escolho n√∫mero de clusters: K = 3

### 2. Escolho aleatoriamente 3 pontos 

Esses pontos ser√£o os centroides dos cluster

````python
plt.figure(figsize=(20,10))
plt.scatter(x, y, c = 'grey', s = 250, zorder = 2)
plt.scatter(x[1], y[1], c = 'blue', s = 1250, marker = 'X', zorder = 2)
plt.scatter(x[3], y[3], c = 'green', s = 1250, marker = 'X', zorder = 2)
plt.scatter(x[6], y[6], c = 'orange', s = 1250, marker = 'X', zorder = 2)
plt.ylim(-0.2,0.3)
plt.axhline(c = 'black', zorder = 1)
plt.xticks([]);
plt.yticks([]);
plt.annotate('', xy = (2,0.03), xytext = (3,0.23),arrowprops=dict(facecolor='black'))
plt.annotate('', xy = (4.5,0.03), xytext = (3,0.23),arrowprops=dict(facecolor='black'))
plt.annotate('', xy = (11,0.03), xytext = (3,0.23),arrowprops=dict(facecolor='black'))
plt.axis('off');
plt.text(3, 0.25, 'Estes s√£o os centroides',
         {'color': 'k', 'fontsize': 24, 'ha': 'center', 'va': 'center'});
````

### 3. Dist√¢ncia do primeiro ponto aos tr√™s centroides iniciais

````python
plt.figure(figsize=(20,10))
plt.scatter(x, y, c = 'grey', s = 250, zorder = 2)
plt.scatter(x[1], y[1], c = 'blue', s = 1250, marker = 'X', zorder = 2)
plt.scatter(x[3], y[3], c = 'green', s = 1250, marker = 'X', zorder = 2)
plt.scatter(x[6], y[6], c = 'orange', s = 1250, marker = 'X', zorder = 2)
plt.ylim(-0.2,0.3)
plt.xlim(0,18)
plt.axhline(c = 'black', zorder = 1)
plt.axis('off');

plt.hlines(y=.1, xmin=x[0], xmax=x[1],linewidth=8, color='blue')
plt.hlines(y=.15, xmin=x[0], xmax=x[3],linewidth=8, color='green')
plt.hlines(y=.20, xmin=x[0], xmax=x[6],linewidth=8, color='orange');
````

- as linhas coloridas representam a dist√¢ncia at√© o centro do cluster, representado pelo simbolo X da mesma cor.

### 4. Associar o primeiro ponto ao cluster mais pr√≥ximo

Menor dist√¢ncia at√© o centroide

````python
plt.figure(figsize=(20,10))
plt.scatter(x, y, c = 'grey', s = 250, zorder = 2)
plt.scatter(x[1], y[1], c = 'blue', s = 1250, marker = 'X', zorder = 2)
plt.scatter(x[3], y[3], c = 'green', s = 1250, marker = 'X', zorder = 2)
plt.scatter(x[6], y[6], c = 'orange', s = 1250, marker = 'X', zorder = 2)
plt.ylim(-0.2,0.3)
plt.xlim(0,18)
plt.axhline(c = 'black', zorder = 1)
plt.axis('off');

# associando primeiro ponto ao cluster azul
plt.scatter(x[0], y[0], c = 'blue', s = 250, zorder = 2);
````

- Vamos fazer a clusteriza√ß√£o usando o sklearn

## 1. N√∫mero de clusters k = 3, sele√ß√£o dos centros pelo sklearn

- Usando o par√¢metro ```init = 'k-means++‚Äô``` que seleciona o centro dos clusters de froma a acelerar a convergencia. 

````python
from sklearn.cluster import KMeans

random_state = 42
X = pd.DataFrame(x)


y_pred = KMeans(n_clusters=3,n_init=10,
                random_state=random_state).fit_predict(X)


plt.figure(figsize=(20, 10))
plt.axhline(c = 'black', zorder = 1)
plt.axis('off');

colors = np.array(['orange', 'blue', 'green']);

plt.scatter(x, y, color = colors[y_pred], s = 250, zorder = 2);
````

- Cluster encontrados como esperado, da mesma forma que dividimos intuitivamento no in√≠cio

## 2. N√∫mero de clusters k = 3, escolhemos o centro dos clusters

- Usando o par√¢metro init = um array com os centros 

````python
X = pd.DataFrame(x)

centro= np.array([[2],[4.5],[11]])


y_pred = KMeans(n_clusters=3,n_init=10, init = centro,
                random_state=random_state).fit_predict(X);


plt.figure(figsize=(20, 10))
plt.axhline(c = 'black', zorder = 1)
plt.axis('off');

colors = np.array(['orange', 'blue', 'green'])


plt.scatter(x, y, color = colors[y_pred], s = 250, zorder = 2);
````

- resultado diferente do anterior

- √© igual ao que fizemos manualmente

- veja a mensagem que executou o Kmeans apenas uma vez
